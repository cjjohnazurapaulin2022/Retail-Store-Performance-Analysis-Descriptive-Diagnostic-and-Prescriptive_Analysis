---

## Project Description

A retail business operating 150 stores across four regions — North, South, 
East, and West — needed to understand why performance varied significantly 
across its Mall, Standalone, and Online store formats. Despite similar 
product offerings, stores showed wide gaps in monthly sales, profit margins, 
and marketing returns with no clear explanation.

This analysis applies a full three-layer analytical framework across 
10 business variables to identify the root causes of performance gaps 
and deliver actionable investment recommendations.

**What → Why → So What**
- Descriptive analysis revealed the performance landscape across regions 
  and store types
- Diagnostic analysis uncovered the drivers behind sales, marketing 
  inefficiency, and margin compression
- Prescriptive analysis translated findings into concrete budget and 
  operational recommendations

---

## Key Questions Answered

1. Which region and store type generates the highest sales — and why?
2. Is marketing spend actually generating proportional returns?
3. Are high return rates the primary threat to profit margins?
4. Where should the business invest its next $50,000 in marketing budget?

<h2 style="color:skyblue;">
 G1:Why does the top region outperform the others — is it footfall, transaction size, or employees?
</h2>

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1 - correlation of key drivers against monthly sales
drivers = ['customer_footfall', 'avg_transaction', 
           'num_employees', 'marketing_spend']

correlations = cleanedData[drivers + ['monthly_sales']].corr()['monthly_sales'].drop('monthly_sales').sort_values(ascending=False)

print("=== Correlation with Monthly Sales ===")
print(correlations.round(3))

# Step 2 - visualize correlations
fig, axes = plt.subplots(1, 4, figsize=(18, 4))

colors = {'Mall': '#378ADD', 'Standalone': '#1D9E75', 'Online': '#D85A30'}

for i, driver in enumerate(drivers):
    for stype, grp in cleanedData.groupby('store_type'):
        axes[i].scatter(grp[driver], grp['monthly_sales'],
                       label=stype, alpha=0.6,
                       color=colors[stype], s=40)
    axes[i].set_xlabel(driver, fontsize=11)
    axes[i].set_ylabel('monthly_sales' if i == 0 else '', fontsize=11)
    axes[i].set_title(f'r = {correlations[driver]:.3f}', fontsize=11)
    if i == 0:
        axes[i].legend(fontsize=9)

plt.suptitle('What drives monthly sales? — G1 Diagnostic', 
             fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Step 3 - mean values per store type for context
print("\n=== Driver Averages by Store Type ===")
print(cleanedData.groupby('store_type')[drivers].mean().round(2))
```

```python
pd.crosstab(cleanedData['region'], cleanedData['store_type'])
```

```python
# Step 1 - total and average marketing spend per region
marketing_by_region = cleanedData.groupby('region').agg(
    total_marketing_spend  = ('marketing_spend', 'sum'),
    avg_marketing_spend    = ('marketing_spend', 'mean'),
    avg_monthly_sales      = ('monthly_sales',   'mean'),
    store_count            = ('store_id',        'count')
).round(2).sort_values('avg_monthly_sales', ascending=False)

print("=== Marketing Spend vs Sales by Region ===")
print(marketing_by_region)

# Step 2 - marketing efficiency ratio (sales generated per $1 of marketing)
marketing_by_region['sales_per_marketing_dollar'] = (
    marketing_by_region['avg_monthly_sales'] /
    marketing_by_region['avg_marketing_spend']
).round(2)

print("\n=== Marketing Efficiency by Region ===")
print(marketing_by_region[['avg_marketing_spend',
                            'avg_monthly_sales',
                            'sales_per_marketing_dollar']])

# Step 3 - visualize side by side
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

regions_ordered = marketing_by_region.index.tolist()

colors = ['#378ADD' if r == regions_ordered[0] 
          else '#B5D4F4' for r in regions_ordered]

axes[0].bar(regions_ordered,
            marketing_by_region['avg_marketing_spend'],
            color=colors, edgecolor='none')
axes[0].set_title('Avg marketing spend per region', fontsize=11)
axes[0].set_ylabel('USD')

axes[1].bar(regions_ordered,
            marketing_by_region['avg_monthly_sales'],
            color=colors, edgecolor='none')
axes[1].set_title('Avg monthly sales per region', fontsize=11)
axes[1].set_ylabel('USD')

axes[2].bar(regions_ordered,
            marketing_by_region['sales_per_marketing_dollar'],
            color=colors, edgecolor='none')
axes[2].set_title('Sales per $1 of marketing spend', fontsize=11)
axes[2].set_ylabel('USD generated')

plt.suptitle('Does marketing spend explain North\'s sales lead? — G1 Extended',
             fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Step 4 - also check avg footfall per region to confirm footfall story
print("\n=== Avg Customer Footfall by Region ===")
print(cleanedData.groupby('region')['customer_footfall'].mean()
        .round(0).sort_values(ascending=False))
```

# Insight
"North leads in both marketing spend 10,687/month and efficiency (10.07 return per 1 spent). However the efficiency gap between regions is narrow (10.07 vs 9.75), meaning North's sales lead is driven more by spend volume than by superior marketing effectiveness. At 10,000 average spend, a 0.32 efficiency gap translates to roughly $3,200 extra monthly sales per store — meaningful but not a structural advantage."

<h2 style="color:skyblue;">
G2: Is marketing spend actually paying off — or are high-spend stores underperforming?
</h2>


```python
import numpy as np
```

```python
# Step 1 - scatter marketing spend vs monthly sales colored by store type
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

colors = {'Mall': '#378ADD', 'Standalone': '#1D9E75', 'Online': '#D85A30'}

for stype, grp in cleanedData.groupby('store_type'):
    axes[0].scatter(grp['marketing_spend'], grp['monthly_sales'],
                   label=stype, alpha=0.65,
                   color=colors[stype], s=50)

axes[0].set_xlabel('Marketing spend (USD)', fontsize=11)
axes[0].set_ylabel('Monthly sales (USD)', fontsize=11)
axes[0].set_title('Marketing spend vs monthly sales by store type', fontsize=11)
axes[0].legend(fontsize=9)

# Step 2 - add efficiency ratio per store and flag outliers
cleanedData['sales_per_mkt_dollar'] = (
    cleanedData['monthly_sales'] / cleanedData['marketing_spend']
).round(2)

# Flag bottom 15% as inefficient, top 15% as highly efficient
low_thresh  = cleanedData['sales_per_mkt_dollar'].quantile(0.15)
high_thresh = cleanedData['sales_per_mkt_dollar'].quantile(0.85)

cleanedData['mkt_efficiency'] = pd.cut(
    cleanedData['sales_per_mkt_dollar'],
    bins=[-np.inf, low_thresh, high_thresh, np.inf],
    labels=['Inefficient', 'Average', 'Highly efficient']
)

print("=== Marketing Efficiency Distribution ===")
print(cleanedData['mkt_efficiency'].value_counts())

print("\n=== Efficiency by Store Type ===")
print(cleanedData.groupby(
        ['store_type', 'mkt_efficiency'],
        observed=False
      )
      .size()
      .unstack(fill_value=0))

# Step 3 - avg sales per marketing dollar by store type
efficiency_by_type = cleanedData.groupby('store_type').agg(
    avg_marketing_spend    = ('marketing_spend',      'mean'),
    avg_monthly_sales      = ('monthly_sales',        'mean'),
    avg_sales_per_dollar   = ('sales_per_mkt_dollar', 'mean'),
    inefficient_count      = ('mkt_efficiency',
                               lambda x: (x == 'Inefficient').sum()),
    efficient_count        = ('mkt_efficiency',
                               lambda x: (x == 'Highly efficient').sum())
).round(2)

print("\n=== Full Efficiency Summary by Store Type ===")
print(efficiency_by_type)

# Step 4 - visualize efficiency distribution per store type
cleanedData.boxplot(column='sales_per_mkt_dollar',
           by='store_type', ax=axes[1],
           patch_artist=True)

axes[1].set_title('Marketing efficiency distribution by store type', fontsize=11)
axes[1].set_xlabel('Store type', fontsize=11)
axes[1].set_ylabel('Sales per $1 marketing spend', fontsize=11)
plt.suptitle('')
plt.tight_layout()
plt.show()

# Step 5 - who are the inefficient high spenders?
print("\n=== High Spend, Low Efficiency Stores (Top 10 worst) ===")
print(cleanedData.nsmallest(10, 'sales_per_mkt_dollar')
        [['store_id', 'region', 'store_type',
          'marketing_spend', 'monthly_sales',
          'sales_per_mkt_dollar']].to_string(index=False))
```

<h2 style="color:skyblue;">
Is Marketing Spend Actually Paying Off?
</h2>

### What the data shows
Overall, marketing spend has a moderate positive correlation with monthly sales (**r = 0.608**), suggesting it contributes to revenue but is not the dominant driver. Out of 150 stores:

- **104 stores (69%)** operate at average efficiency  
- **23 stores (15%)** are highly efficient  
- **23 stores (15%)** are inefficient  

### Which store type converts marketing best?

| Store Type | Avg Marketing Spend | Avg Monthly Sales | Sales per \$1 Spent |
|---|---:|---:|---:|
| Online | \$9,540 | \$95,492 | \$10.99 |
| Standalone | \$8,451 | \$84,444 | \$10.67 |
| Mall | \$11,176 | \$108,033 | \$10.35 |

Online stores generate the highest return per marketing dollar (**\$10.99**) despite spending the least, making them the most efficient channel.

Mall stores spend the most (**\$11,176/month**) but deliver the lowest efficiency (`**\$10.35 `per \$1 spent**), suggesting diminishing returns at higher spending levels.

### The inefficiency problem
Online stores have the highest count of inefficient stores (**9 out of 59**), but also the highest count of highly efficient stores (**11 out of 59**). This makes Online a **high-variance channel** — it produces both the strongest and weakest performers.

Mall stores show the opposite pattern:  
- High marketing spend  
- Lower efficiency ceiling  
- Fewer breakout performers  

The 10 worst-performing stores span all three store types and all four regions, suggesting inefficiency is **not structural**. Instead, performance issues are likely driven by **store-specific execution problems** rather than weaknesses in the business model itself.

### Why it matters
Marketing spend is paying off overall, but returns are uneven across store formats.

The business is not getting consistent returns from Mall stores despite those stores receiving the largest budgets. Meanwhile, Online stores generate stronger returns on lower spend levels, although with higher variability and risk.

### This means...
Marketing investment is producing positive returns across all store types, but Mall stores remain the least efficient channel despite receiving the highest budgets.

This suggests the business may be:
- **Over-investing** in its most expensive format (Mall stores)
- **Under-investing** in its most efficient format (Online stores)

A more optimized allocation strategy could improve overall marketing ROI while reducing wasted spend.

<h2 style="color: #378ADD;"> G3:Are high return rates killing profit margins — or is something else responsible?</h2>




```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(10,8))
plt.scatter(cleanedData["return_rate"],cleanedData["profit_margin"],
    color="blue",
    alpha=0.6,
    s=80)

# Trend line
z = np.polyfit(cleanedData["return_rate"],cleanedData["profit_margin"], 1)
p = np.poly1d(z)

plt.plot(
    cleanedData["return_rate"],
    p(cleanedData["return_rate"]),
    color="red",
    linewidth=3 )
        
plt.xlabel("Return Rate",fontsize=12)
plt.ylabel("profit_margin",fontsize=12)
plt.title("Relationship Between Return Rate and Profit Margin Across All Stores", fontsize=14)
plt.grid()
plt.show
```

```python
# Step 1 - scatter return rate vs profit margin
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

colors = {'Mall': '#378ADD', 'Standalone': '#1D9E75', 'Online': '#D85A30'}

for stype, grp in cleanedData.groupby('store_type'):
    axes[0].scatter(grp['return_rate'], grp['profit_margin'],
                   label=stype, alpha=0.65,
                   color=colors[stype], s=50)

axes[0].set_xlabel('Return rate', fontsize=11)
axes[0].set_ylabel('Profit margin', fontsize=11)
axes[0].set_title('Return rate vs profit margin by store type', fontsize=11)
axes[0].legend(fontsize=9)

# Step 2 - boxplot profit margin by store type
cleanedData.boxplot(column='profit_margin', by='store_type',
           ax=axes[1], patch_artist=True)
axes[1].set_title('Profit margin distribution by store type', fontsize=11)
axes[1].set_xlabel('Store type', fontsize=11)
axes[1].set_ylabel('Profit margin', fontsize=11)
plt.suptitle('')

plt.suptitle('Do return rates kill profit margins? — G3 Diagnostic',
             fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Step 3 - correlation between return rate and profit margin
corr_return_profit = cleanedData['return_rate'].corr(cleanedData['profit_margin'])
print(f"=== Correlation: Return Rate vs Profit Margin ===")
print(f"r = {corr_return_profit:.3f}")

# Step 4 - avg return rate and profit margin by store type
print("\n=== Return Rate vs Profit Margin by Store Type ===")
summary = cleanedData.groupby('store_type').agg(
    avg_return_rate  = ('return_rate',    'mean'),
    avg_profit_margin= ('profit_margin',  'mean'),
    max_return_rate  = ('return_rate',    'max'),
    min_profit_margin= ('profit_margin',  'min')
).round(4)
print(summary)

# Step 5 - does marketing spend compound the margin problem?
print("\n=== Combined Pressure on Profit Margin ===")
cleanedData['margin_pressure'] = (
    cleanedData['return_rate'] + 
    (cleanedData['marketing_spend'] / cleanedData['monthly_sales'])
).round(4)

print(cleanedData.groupby('store_type')['margin_pressure']
        .mean().round(4)
        .sort_values(ascending=False))
```

<h2 style="color:skyblue;">
Are High Return Rates Killing Profit Margins?
</h2>

### What the data shows
Return rate has a moderate to strong negative correlation with profit 
margin (r = -0.621), meaning higher returns consistently compress margins 
across the dataset. This is not a weak relationship — return rate is one 
of the most significant margin drivers in this analysis.

### Return rate vs profit margin by store type
| Store Type | Avg Return Rate | Avg Profit Margin | Min Profit Margin |
|------------|----------------|-------------------|-------------------|
| Mall       | 8.25%          | 24.09%            | 16.97%            |
| Standalone | 9.95%          | 23.28%            | 17.65%            |
| Online     | 17.50%         | 19.62%            | 10.54%            |

Online stores carry more than double the return rate of Mall stores 
(17.5% vs 8.25%), which directly explains their lower average profit 
margin (19.62% vs 24.09%). The damage is real and measurable.

### Why Online margins survive despite high returns
Online stores have the highest combined margin pressure score (0.2743) 
from both return rate and marketing spend ratio combined. However, Online 
stores partially offset return damage through significantly lower marketing 
spend (`$ 9,540/month vs Mall's  $ 11,176/month`). The marketing savings act 
as a buffer against the cost of high returns.

This explains the D3 observation — Online and Mall margins appear 
similar on the surface, but they are being compressed by completely 
different forces. Mall margins are threatened by high marketing spend. 
Online margins are threatened by high return rates. They arrive at 
similar outcomes through opposite problems.

### The most dangerous stores
Stores facing BOTH high return rates AND high marketing spend ratios 
simultaneously face the worst combined margin pressure. These stores 
have no offsetting buffer — every return costs margin and every 
marketing dollar compounds the damage.

### This means...
Return rates are a significant and measurable margin killer (r = -0.621), 
but Online stores partially survive high returns through lower marketing 
spend — meaning the real threat to profitability is not return rate or 
marketing spend alone, but stores that suffer both pressures at the same 
time with no offsetting advantage.

<h2 style="color: skyblue;">
P1 — Where Should the Business Invest by Store Type?
</h2>

### Recommendation Summary
| Store Type | Action        | Reason |
|-------------|---------------|--------|
| Mall        | Invest        | Highest sales ($108,033) and margin (24.09%) |
| Online      | Invest + Fix  | Best marketing efficiency but high return risk |
| Standalone  | Maintain      | Balanced performance, no strong advantage |

### Mall Stores — Invest
Mall stores lead in average monthly sales (`$108,033`) and profit margin 
(24.09%), while also showing the lowest margin pressure score (0.1857). 
Although they use the highest marketing budget (`$11,176/month`), they still 
generate strong efficiency at `$10.35` per $1 spent. Mall stores offer the 
best balance of stability and profitability.

### Online Stores — Invest with Conditions
Online stores achieve the highest marketing efficiency (`$10.99` per $1 spent) 
with the lowest marketing budget (`$9,540/month`). However, their high return 
rate (17.5%) reduces margins to 19.62%, the weakest among all formats. 
Investment should focus on reducing returns before scaling further.

### Standalone Stores — Maintain
Standalone stores remain consistent but do not lead in any metric. With 
average sales of `$84,444`, profit margins of 23.28%, and marketing efficiency 
of `$10.67` per dollar, they provide stable but limited growth potential. 
Maintain current investment levels while prioritizing Mall and Online stores.

### This means...
The business should prioritize Mall stores for stable growth and Online 
stores for efficiency gains, while improving Online return rates before 
expansion.

<h2 style="color: skyblue;">
P2 — Where Should the $50,000 Marketing Budget Go?
</h2>

### Recommended Allocation
| Channel | Allocation | Rationale |
|----------|------------|------------|
| Mall | `$30,000 `| Highest sales and lowest margin risk |
| Online | `$15,000` | Best efficiency with higher upside |
| Standalone | `$5,000` | Maintain presence without scaling |

### Mall — `$30,000` (60%)
Mall stores generate `$10.35` in sales per `$1` of marketing spend and 
show the lowest margin pressure score (0.1857). A `$30,000` increase is 
expected to generate about $310,500 in additional sales. They also have 
fewer inefficient stores (7 of 49), reducing investment risk.

### Online — `$15,000` (30%)
Online stores deliver the highest marketing efficiency at `$10.99` per 
$1 spent, producing an estimated `$164,850` in additional sales from a 
`$15,000` increase. However, high return rates (17.5%) and more 
inefficient stores (9 of 59) make this a higher-risk channel. The 
allocation captures growth potential while limiting exposure.

### Standalone — `$5,000` (10%)
Standalone stores show stable but average performance with no major 
efficiency advantage. The $5,000 allocation is intended to maintain 
operations rather than drive expansion.

### North Region Priority
Mall and Online investments should prioritize North region stores. 
North leads in average sales (`$107,599`), marketing efficiency 
($10.07 per dollar), and customer footfall (4,937 visitors monthly), 
making it the strongest region for incremental returns.

### This means...
A 60/30/10 split across Mall, Online, and Standalone stores — focused 
on North region locations — is projected to generate roughly $475,000 
in additional monthly sales while balancing stable returns and growth 
opportunities.
